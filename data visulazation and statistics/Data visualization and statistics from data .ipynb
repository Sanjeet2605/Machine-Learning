```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("pima-indians-diabetes.csv")
print("Question 2.")
print()
print()
figure, axis = plt.subplots(2, 4)
# between age and pregs
axis[0, 0].scatter(data["Age"], data["pregs"])
axis[0, 0].set_title("b/w Age and pregs")

# between age and plas
axis[0, 1].scatter(data["Age"], data["plas"])
axis[0, 1].set_title("b/w Age and plas")

# between age and pres
axis[0, 2].scatter(data["Age"], data["pres"])
axis[0, 2].set_title("b/w Age and pres")

# between age and skin
axis[0, 3].scatter(data["Age"], data["skin"])
axis[0, 3].set_title("b/w Age and skin")

# between age and BMI
axis[1, 0].scatter(data["Age"], data["BMI"])
axis[1, 0].set_title("b/w Age and BMI")

# between age and test
axis[1, 1].scatter(data["Age"], data["test"])
axis[1, 1].set_title("b/w Age and test")

# between age and pedi
axis[1, 2].scatter(data["Age"], data["pedi"])
axis[1, 2].set_title("b/w Age and pedi")

# Combine all the operations and display
plt.show()

figure, axis = plt.subplots(2, 4)
# between BMI and pregs
axis[0, 0].scatter(data["BMI"], data["pregs"])
axis[0, 0].set_title("b/w BMI and pregs")

# between BMI and plas
axis[0, 1].scatter(data["BMI"], data["plas"])
axis[0, 1].set_title("b/w BMI and plas")

# between BMI and pres
axis[0, 2].scatter(data["BMI"], data["pres"])
axis[0, 2].set_title("b/w Age and pres")

# between BMI and skin
axis[0, 3].scatter(data["BMI"], data["skin"])
axis[0, 3].set_title("b/w Age and skin")

# between BMI and Age
axis[1, 0].scatter(data["BMI"], data["Age"])
axis[1, 0].set_title("b/w Age and BMI")

# between BMI and test
axis[1, 1].scatter(data["BMI"], data["test"])
axis[1, 1].set_title("b/w Age and test")

# between BMI and pedi
axis[1, 2].scatter(data["BMI"], data["pedi"])
axis[1, 2].set_title("b/w Age and pedi")

# Combine all the operations and display
plt.show()
```
